```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\yasha\Desktop\Ashish\sem 6\DSBDA\DSBDA Lab Datasets\iris.csv")
df
```

```python
# line graph
x=df['sepal_length']
y=df['sepal_width']
plt.xlabel("sepal lenght")
plt.ylabel("sepal width")
plt.title("sepal")
plt.plot(x,y)
```

```python
#multi line grapgh
x=df['species']
y=df['sepal_width']
z=df['sepal_length']
plt.plot(x,y,label="sepal width")
plt.plot(x,z,label="sepal lenght")
plt.legend()
plt.show()
```

```python
#histogram
x=np.random.normal(df['petal_width'])
plt.hist(x)
```

```python
#box plot
sns.boxplot(x=df['sepal_length'],notch=True)
```

```python
#heat map
sns.heatmap(df.corr())
```

```python
#scatter plot
x=(df['sepal_length'])
y=(df['sepal_width'])
plt.scatter(x,y,color='red')
```

```python
#pie chart
plt.pie(x=df['petal_length'].value_counts())
plt.show()
```
