```python
import requests  # used to send http reuest act as request and respond protocolo to client and server
import bs4       # pull data from particular page
```

```python
request1 = requests.get("https://www.flipkart.com/vivo-v27-5g-magic-blue-128-gb/p/itm5f209bb952b0f?pid=MOBGNCWFZZUHHAFD&lid=LSTMOBGNCWFZZUHHAFDAJHMFP&marketplace=FLIPKART&store=tyy%2F4io&srno=b_1_1&otracker=clp_bannerads_1_28.bannerAdCard.BANNERADS_A_mobile-phones-store_0NNRHF09GYJC&fm=organic&iid=4fa4ab2e-a9ad-4fe7-b78b-da1acaf14246.MOBGNCWFZZUHHAFD.SEARCH&ppt=clp&ppn=mobile-phones-store&ssid=nm4fzfvwv40000001685285702852")
request1
```

```python
request1.content  # we will get http document or raw data 
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup   3 we will get text from the above document
```

```python
# fetching the all reviews/comments
reviews = soup.findAll('div',{'class' : 't-ZTKy'});
for review in reviews:
    print(review.get_text() + '\n\n')
    
```

```python
# fetching overall rating
ratings = soup.find('div',{'class' : '_2d4LTz'}).get_text()
print(ratings)
```

```python
# fetching all ratings 
indi_ratings = soup.findAll('div',{'class' : '_3LWZlK _1BLPMq'});
for individual in indi_ratings:
    print(individual.get_text() + '\n\n')
```

```python
# fetching customer names
cust_name = soup.findAll('p',{'class' : '_2sc7ZR _2V5EHH'});
for customer_name in cust_name:
    print(customer_name.get_text() + '\n')
```
